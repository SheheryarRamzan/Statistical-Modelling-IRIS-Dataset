```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st

# Define the mean and standard deviation
mean = 7
std = 2

# Define the domain of x (i.e., the range of values for the parameter)
x = np.linspace(0, 15, 100)

# Generate the two prior distributions
dist1 = st.norm(loc=mean, scale=std)
dist2 = st.gamma(a=mean**2/std**2, scale=std**2/mean)

# Plot the Normal prior distribution
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.plot(x, dist1.pdf(x))
ax1.axvline(x=dist1.mean(), color="red", linestyle="--", label='Mean')
ax1.set_xlabel("Parameter Value")
ax1.set_ylabel("Density")
ax1.set_title("Normal Prior Distribution")

# Plot the Gamma prior distribution
ax2.plot(x, dist2.pdf(x))
ax2.axvline(x=dist2.mean(), color="red", linestyle="--")
ax2.set_xlabel("Parameter Value")
ax2.set_ylabel("Density")
ax2.set_title("Gamma Prior Distribution")

# Show the plots
plt.show()

# Get the mean and standard deviation of each distribution
print("Mean and Standard Deviation of Normal Distribution:", dist1.mean(), dist1.std())
print("Mean and Standard Deviation of Gamma Distribution:", dist2.mean(), dist2.std())
```
